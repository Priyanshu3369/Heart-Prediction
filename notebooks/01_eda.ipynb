```python
import pandas as pd
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../data/raw/heart.csv")
sns.countplot(data=df, x='target', palette='Set2')
plt.title("Heart Disease Class Distribution (Target)")
plt.xlabel("Heart Disease (1 = Yes, 0 = No)")
plt.ylabel("Count")
plt.savefig("../outputs/figures/target_distribution.png")
plt.show()
```

```python
plt.figure(figsize=(12, 10))
corr = df.corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", square=True)
plt.title("Correlation Matrix of Heart Disease Features")
plt.tight_layout()
plt.savefig("../outputs/figures/correlation_heatmap.png")
plt.show()
```

```python
numeric_cols = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']

for col in numeric_cols:
    plt.figure()
    sns.histplot(df[col], kde=True, color='teal')
    plt.title(f'Distribution of {col}')
    plt.savefig(f'../outputs/figures/dist_{col}.png')
    plt.show()
```

```python
# Age vs Target
sns.boxplot(data=df, x="target", y="age", palette="Set3")
plt.title("Age vs Target")
plt.savefig("../outputs/figures/age_vs_target.png")
plt.show()

# Cholesterol vs Target
sns.boxplot(data=df, x="target", y="chol", palette="Set3")
plt.title("Cholesterol vs Target")
plt.savefig("../outputs/figures/chol_vs_target.png")
plt.show()
```
